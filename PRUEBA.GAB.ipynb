```python

# Buscar artista por nombre
resultado = sp.search(q="Shakira", type="artist", limit=1)

if resultado['artists']['items']:
    artista = resultado['artists']['items'][0]
    artist_id = artista['id']
    print(" Artista encontrado:", artista['name'])
    print(" ID:", artist_id)
    print(" Géneros:", ", ".join(artista['genres']))
    print(" Perfil Spotify:", artista['external_urls']['spotify'])
    print()

    # Obtener álbumes del artista
    albums = sp.artist_albums(artist_id, album_type='album', limit=50)

    print(" Álbumes lanzados entre 2002 y 2017:\n")

    for album in albums['items']:
        fecha = album['release_date']
        # Algunos álbumes tienen fecha incompleta (solo año o año-mes)
        try:
            anio = int(fecha[:4])
        except:
            continue  # Ignorar fechas inválidas

        if 2002 <= anio <= 2017:
            print(" Álbum:", album['name'])
            print(" Año:", anio)
            print(" ID:", album['id'])
            print(" Spotify:", album['external_urls']['spotify'])
            print()

            # Obtener canciones del álbum
            tracks = sp.album_tracks(album['id'])
            print(" Canciones del álbum:\n")
            for i, track in enumerate(tracks['items'], start=1):
                print(f"   {i}. {track['name']}")
                print("      ID:", track['id'])
                print("       Spotify:", track['external_urls']['spotify'])
            

else:
    print(" No se encontró el artista.")
```
