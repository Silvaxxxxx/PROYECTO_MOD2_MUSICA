```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import csv
import requests

# Configuración
CLIENT_ID = "9beba5ec7f124f4282d3625eecb0339f"
CLIENT_SECRET = "c6501c7f4217420095f22bef5dfed7b9"

# Conectar a Spotify
client_credentials_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
# Configuración del proyecto
años = [2010, 2011, 2012, 2013, 2014, 2015, 2016]
datos = []

print("Extrayendo datos de género metal de Spotify...")

# Buscar canciones y álbumes
for año in años:
    print(f"  Año: {año}")

    for o in range (1,101,50):
            
        
        # Buscar canciones
        try:
            resultados = sp.search(q=f'genre:"metal" year:{año}', type='track,album', limit=50, offset=o)
                
            for cancion in resultados['tracks']['items']:
                if cancion['album']['release_date']:
                    año_lanzamiento = int(cancion['album']['release_date'][:4])
                        
                    # Solo guardar si está en nuestro rango
                    if año_lanzamiento in años:
                        datos.append({
                            'artista': cancion['artists'][0]['name'],
                            'genero_musical': "metal",
                            'tipo': 'cancion',
                            'nombre': cancion['name'],
                            'año_lanzamiento': año_lanzamiento,
                            'id': cancion['id'],
                            'popularidad': cancion['popularity']
                        })
        except Exception as e:
            print(f"    Error buscando canciones para metal {año}: {e}")
    
    # Mostrar estadísticas básicas
canciones = sum(1 for item in datos if item['tipo'] == 'cancion')
albumes = sum(1 for item in datos if item['tipo'] == 'album')

print(f"Canciones: {canciones}")
print(f"Álbumes: {albumes}")

print("\nTotal:")

count = sum(1 for item in datos if item['genero_musical'] == "metal")
print(f"Metal: {count}")

# Guardar en CSV
print(f"\nGuardando {len(datos)} elementos en CSV...")

with open('spotify_datos_metal.csv', 'w', newline='', encoding='utf-8') as archivo:
    writer = csv.DictWriter(archivo, fieldnames=['artista', 'genero_musical', 'tipo', 'nombre', 'año_lanzamiento', 'id', 'popularidad'])
    writer.writeheader()
    writer.writerows(datos)

print("¡Extracción completada!")
print(f"Total de elementos: {len(datos)}")

# Mostrar estadísticas básicas
canciones = sum(1 for item in datos if item['tipo'] == 'cancion')
albumes = sum(1 for item in datos if item['tipo'] == 'album')

print(f"Canciones: {canciones}")
print(f"Álbumes: {albumes}")

print("\nPor género:")
count = sum(1 for item in datos if item['genero_musical'] == "metal")
print(f"Metal: {count}")
```
